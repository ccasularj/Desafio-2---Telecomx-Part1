## Extração dos dados

Os dados foram obtidos a partir de uma API disponibilizada no GitHub em formato JSON.

Nesta etapa, os dados são carregados diretamente utilizando Python e convertidos em um DataFrame do Pandas para facilitar a manipulação e análise.

```python
import pandas as pd

url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science/main/TelecomX_Data.json"

df = pd.read_json(url)

df.head()
```

#🔧 Transformação

## Estrutura dos dados

Após a importação dos dados, é importante entender a estrutura do dataset.

Nesta etapa verificamos:

- número de linhas e colunas

## Verificação de inconsistências

Nesta etapa realizamos uma análise inicial para identificar possíveis problemas nos dados que possam afetar as análises posteriores.

Foram verificados:

- valores ausentes
- registros duplicados
- tipos de dados incorretos
- inconsistências nas categorias

```python
df.isnull().sum()
df.duplicated().sum()
df.dtypes
```

## Tratamento das inconsistências

Após identificar possíveis problemas nos dados, realizamos algumas correções para garantir a consistência do dataset.

As principais ações realizadas foram:

- conversão de tipos de dados
- tratamento de valores ausentes
- remoção de duplicados

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
df.isnull().sum()
df = df.dropna()
df = df.drop_duplicates()

```

## Criação da variável Contas_Diarias

Para enriquecer a análise, foi criada a variável **Contas_Diarias**, que representa o valor médio diário pago pelo cliente.

Essa variável foi calculada a partir da cobrança mensal.

```python
df["Contas_Diarias"] = df["MonthlyCharges"] / 30

df[["MonthlyCharges", "Contas_Diarias"]].head()
```

## Distribuição de evasão

Nesta etapa analisamos como a variável **Churn** está distribuída entre os clientes.

Isso permite entender qual a proporção de clientes que cancelaram o serviço em comparação aos que permaneceram.

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df["Churn"].value_counts()
```

```python
sns.countplot(x="Churn", data=df)

plt.title("Distribuição de Evasão de Clientes")

plt.show()
```

```python
df["Churn"].value_counts(normalize=True).plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(6,6)
)

plt.title("Proporção de Clientes que Cancelaram")

plt.ylabel("")

plt.show()
```

## Evasão por variáveis categóricas

Nesta etapa analisamos como a evasão se distribui de acordo com variáveis categóricas como:

- gênero
- tipo de contrato
- método de pagamento
- serviços contratados

Essa análise ajuda a identificar perfis de clientes com maior tendência a cancelar o serviço.

```python
categorical_columns = [
"gender",
"Partner",
"Dependents",
"InternetService",
"Contract",
"PaymentMethod"
]
```

```python
for col in categorical_columns:

    plt.figure(figsize=(6,4))

    sns.countplot(x=col, hue="Churn", data=df)

    plt.title(f"Evasão por {col}")

    plt.xticks(rotation=45)

    plt.show()
```

## Evasão por variáveis numéricas

Nesta etapa analisamos como variáveis numéricas se comportam em relação à evasão de clientes.

Entre as variáveis analisadas estão:

- tempo de contrato (tenure)
- cobrança mensal (MonthlyCharges)
- valor total pago (TotalCharges)

```python
numerical_cols = [
"tenure",
"MonthlyCharges",
"TotalCharges",
"Contas_Diarias"
]
```

```python
for col in numerical_cols:

    plt.figure(figsize=(6,4))

    sns.boxplot(x="Churn", y=col, data=df)

    plt.title(f"{col} vs Churn")

    plt.show()
```

# Relatório Final

## Introdução

O objetivo desta análise foi compreender os fatores que influenciam a evasão de clientes em uma empresa de telecomunicações.

## Limpeza e Tratamento de Dados

Os dados foram extraídos de uma API em formato JSON e convertidos em um DataFrame utilizando Pandas.

Durante o processo de limpeza foram realizadas as seguintes etapas:

- verificação de valores ausentes
- conversão de tipos de dados

## Análise Exploratória

A análise exploratória revelou alguns padrões importantes:

- clientes com contratos mensais apresentam maior taxa de cancelamento
- clientes com menor tempo de relacionamento tendem a cancelar mais
- cobranças mensais mais altas estão associadas a maior churn

## Conclusões

Os resultados indicam que fatores como tipo de contrato, tempo de permanência e valor da cobrança mensal influenciam significativamente a evasão de clientes.

## Recomendações

Com base nos resultados, algumas estratégias podem ser adotadas:
